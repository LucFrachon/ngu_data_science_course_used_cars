# Used Cars - Complete Example

In this notebook, we take everything we learned in _Week 6's "Used Cars - Load and Inspect"_, _Week 7's "User Cars - EDA"_, and _Week 8's "Cleaning and Formatting" (section "worked example")_ to write the data preprocessing pipeline, feature engineering and selection process, and the modelling operations.

## Load and preprocess data

```python
import numpy as np, pandas as pd
```

```python
def load_and_split(filepath, val_ratio, test_ratio, seed=0):
    """
    Load raw data file and split into train, validation, and test sets. 
    val_ratio: Ratio in [0, 1], in proportion of the raw data size.
    test_ratio: Ratio in [0, 1], in proportion of the raw data size.
    val_ratio + test_ratio must be > 0 and < 1.
    """
    assert 0 < val_ratio + test_ratio < 1
    # 1. Load the translated and clean dataset
    cars = pd.read_csv(
        filepath,
        parse_dates=[0, 16, 19],
        dtype={
            'seller': 'category', 
            'offerType': 'category',
            'abtest': 'category' , 
            'vehicleType': 'category',
            'monthOfRegistration': 'category',
            'gearbox': 'category',
            'model': 'category',
            'fuelType': 'category',
            'brand': 'category',
            'notRepairedDamage': 'category',
            'postalCode': 'str'
        },
    )

    # 2. Split into train, validation, and test sets
    test_idx = np.random.choice(
        cars.index, 
        size=round(len(cars) * test_ratio), 
        replace=False
    )
    cars_test = cars.loc[test_idx]
    cars_tr_val = cars.loc[cars.index.difference(cars_test.index)]
    val_idx = np.random.choice(
        cars_tr_val.index,
        size=round(len(cars) * val_ratio),
        replace=False
    )
    cars_val = cars_tr_val.loc[val_idx]
    cars_train = cars_tr_val.loc[cars_tr_val.index.difference(cars_val.index)]
    
    print(f"Generated 3 splits:\n"
          f"- Train of length {len(cars_train)},\n"
          f"- Validation of length {len(cars_val)},\n"
          f"- Test of length {len(cars_test)}.")
       
    return cars_train, cars_val, cars_test
```

```python
import warnings

def remove_features(data):
    data = data.copy()
    if len(data.loc[(data.seller != 'private') | (data.offerType != 'selling')]) > 0:
        warnings.warn(f"Your data contains examples of professional seller, or of buyer's add."
                      f" These will be removed.")
        data = data.loc[(data.seller=='private') & (data.offerType=='selling')]
    data.drop(['dateCrawled', 'dateCreated', 'lastSeen', 
               'name', 'seller', 'offerType', 'abtest', 'nrOfPictures'], 
              axis=1, inplace=True)
    return data
```

```python
def cleanup(data, split='train'):
    """
    Perform initial cleanup of `data`.
    split: Specify whether "train", "valid", or "test"
    """
    assert split in ['train', 'valid', 'test']
    data = data.copy()
    if split == 'train':
        data.drop_duplicates(inplace=True)
    # Make hidden NaNs explicit
    data.monthOfRegistration.replace('0', np.nan, inplace=True)
    # Clean up the categories in monthOfRegistration and reorder properly
    data.monthOfRegistration = data.monthOfRegistration.cat.remove_unused_categories()
    data.monthOfRegistration = data.monthOfRegistration.cat.reorder_categories(
        [str(s) for s in range(1, 13)],
        ordered=True,
    )
    data.price.replace(0, np.nan, inplace=True)
    data.loc[data.powerPS <= 0, 'powerPS'] = np.nan
    data.loc[:, 'notRepairedDamage'] = np.where(data['notRepairedDamage'] == 'yes', 1, 0)
    return data    
```

 ## Feature and Label Sanitisation
 Here, we are going to apply the things we learned in EDA to make the features and the label cleaner, in particular by handling outliers. We are also splitting the features from the labels.

```python
def make_labels(data, log_transform=True, label_bounds=(100, 100000)):
    """
    Splits the features from the labels, and removes NA label rows and rows where the label
    has extreme values (as defined in `label_bounds`). Optionally applies a logarithmic 
    transformation (base 10) to the label values.
    :Returns: Feature dataframe X, label vector y.
    """
    data = data.copy()
    # Remove rows with no price information
    data = data[~data['price'].isnull()]
    # Remove outliers
    data = data.query(f"(price >= {label_bounds[0]}) & (price <= {label_bounds[1]})")
    if log_transform:
        y = np.log10(data['price'])
        X = data.drop('price', axis=1)
    else:
        y = data['price']
        X = data.drop('price', axis=1)
    return X, y 
```

```python
def outliers_to_nans(X, power_bounds=(25, 650), year_bounds=(1960, 2016), recent_km=10000):
    """
    Sets to NA observations that are considered as outliers. This allows to then handle them
    later within the NA imputation mechanism.
    :power_bounds, year_bounds: Lower and upper limits for PowerPS and YearOfRegistration respectively. 
    Any value more extreme than these are considered as an outlier and set to explicit np.nan.
    :recent_km: Highest accepted value for vehicles registered in the year that constitutes the
    upper bound in `year_bounds` (i.e. 2016 by default). Any vehicle from that year with more kilometers
    will be considered as an outlier and `kilometer` will be set to explicit np.nan.
    :Returns: Dataframe of features X with outlier corrections as per the above.
    """
    X = X.copy()
    # Set to NA all vehicules with too low or too high powerPS. We add a binary column 
    # to record which examples used to be outliers
    X['powerPS_out'] = np.where(
        (X.powerPS < power_bounds[0]) | (X.powerPS > power_bounds[1]), 
        1, 
        0
    )
    X.loc[(X.powerPS < power_bounds[0]) | (X.powerPS > power_bounds[1]), 
             'powerPS'] = np.nan
    # Set to NA all vehicles with too low or too high YearOfRegistration. We add a binary column 
    # to record which examples used to be outliers
    X['yearOfRegistration_out'] = np.where(
        (X['yearOfRegistration'] < year_bounds[0]) | (X['yearOfRegistration'] > year_bounds[1]),
        1,
        0
    )
    X.loc[
        (X['yearOfRegistration'] < year_bounds[0]) |
        (X['yearOfRegistration'] > year_bounds[1]),
        'yearOfRegistration'
    ] = np.nan
    # Set to NA recent vehicles (latest accepted year in data) with high km
    X.loc[
        (X['yearOfRegistration'] == year_bounds[1]) &
        (X['kilometer'] > recent_km),
        'yearOfRegistration_out'
    ] = 1
    X.loc[
        (X['yearOfRegistration'] == year_bounds[1]) &
        (X['kilometer'] > recent_km),
        'yearOfRegistration'
    ] = np.nan
    return X
```

## Feature Engineering

Again, we are going to write functions to create usable and informative features. During EDA, we realised that prices of cars aged less than 20 years decreased with age, but the opposite was happening to cars older than 20 years. We hypothesised that after 20 years, a vehicle became a vintage collectible item whose value increased with age. So we are going to add a `vintage` flag that will help any linear models that we want to fit.

```python
def add_vintage_interaction(data, pivot_year = 1996, create_product = True):
    """
    Creates a binary feature `vintage` that takes 1 if `yearOfRegistration` <= `pivot_year`, 
    and 0 otherwise.
    If `create_product` == True, also adds a feature equal to `vintage` * `yearOfRegistration` (useful
    in linear regression).
    :return: The input dataframe with additional columns as specified above.
    """
    data = data.copy()
    data['vintage'] = np.where(data['yearOfRegistration'] <= pivot_year, 1, 0)
    if create_product:
        data['vintage_x_year'] = data['vintage'] * data['yearOfRegistration']
    return data
```

## Baseline Model
At this stage, we are almost ready to fit our baseline model. There is still a bit of feature engineering to do with `postalCode`. It's currently a string variable that has far too many categories to one-hot encode. So we must reduce the number of categories somehow. To keep things simple, let's just retain the first two characters. The reason why we are doing this now and not earlier (in the Feature Engineering section) is that we will want to experiement with different encoding methods on `postalCode`. Here, we just use the simplest method. 

```python
def transform_postcode(X, n_digits=2):
    X = X.copy()
    X['postalCode'] = X['postalCode'].str[:n_digits]
    return X
```

Then, we will need to do three more things before we can finally fit our model: impute missing values, scale all quantitative variables, and one-hot encode categorical variables.

To define what imputation method to use with each variable, let's create a configuration dictionary that we can pass to our function. The function also offers the option of adding a binary flag for each imputed variable that records which observations were missing.

__Warning__: We must be cautious here. To avoid data leakage from the validation and test sets, we must learn the values to impute from the training set only, and apply them to all three splits. The best way to do this is to subclass sklearn's TransformerMixin class, which we will do later. For now, let's do it manually but bear in mind that this presents more risks of accidental data leakage.

```python
NA_METHODS = dict(
    vehicleType         ='mode',
    yearOfRegistration  ='median',
    gearbox             ='mode',
    powerPS             ='median',
    model               ='mode',
    kilometer           ='median',
    monthOfRegistration ='mode',
    fuelType            ='mode',
    brand               ='mode',
    notRepairedDamage   ='mode',
    postalCode          ='zero'   
)
```

```python
def learn_impute_values(X_train, methods_dict, zero_n_chars = 2):
    """
    Learn the imputation values from the training set. Do not use on the validation or test sets!
    X_train: Training set (features only)
    methods_dict: Configuration dictionary mapping each feature to the imputation method to use on it.
    zero_n_chars: When using method `zero` on a string feature, how many characters to impute. 
    Defaults to 2 (i.e. `00`).
    :return: A dictionary mapping each feature to the imputation value to use, learned from the 
    training set.
    """
    X = X_train.copy()
    values = dict()
    for var in X.columns:
        # Default method is 'mode' (for variables that are not in methods_dict):
        if var not in methods_dict:
            values[var] = X[var].mode()[0]
        elif methods_dict[var] == 'mode':
            values[var] = X[var].mode()[0]
        elif methods_dict[var] == 'median':
            values[var] = X[var].median()
        elif methods_dict[var] == 'mean':
            values[var] = X[var].mean()
        elif methods_dict[var] == 'zero':
            values[var] = '0' * zero_n_chars
        elif methods_dict[var] == 'skip':
            continue
        else:
            raise ValueError("Methods can only be mode, median, mean, zero, or skip.")
    return values
```

Test:

```python
impute_values = learn_impute_values(x_temp, NA_METHODS)
impute_values
```

```python
def impute_NA(X, values_dict, na_flag = True):
    X = X.copy()
    for var in X.columns:
        # Do not create binary flags for variables that are already flags created by us
        if na_flag and (var[-4:] != '_out') and (var[-3:] != '_NA'):
            X[var + '_NA'] = np.where(X[var].isnull(), 1, 0)
        X[var] = X[var].fillna(values_dict[var])
        
    return X
```

For scaling, just like for imputation, we need to learn the scaling parameters from the training set. So we first create a function to learn these parameters. Again, using a dedicated subclass of TransformerMixing would allow us to do this more elegantly, but we'll get to that later.

```python
SCALING_METHODS = dict(
    yearOfRegistration   = 'normalize',
    powerPS              = 'standardize',
    kilometer            = 'normalize',
#     monthOfRegistration  = 'normalize',
)
```

```python
def learn_scaling_params(X, methods):
    X = X.copy()
    params = dict()
    for var, method in methods.items():
        if method == 'normalize':
            params[var] = {'min': X[var].min(), 'max': X[var].max()}
        elif method == 'standardize':
            params[var] = {'mean': X[var].mean(), 'stdev': X[var].std()}
        else:
            raise ValueError("Methods can only be normalize and standardise.")
    return params
```

Test:

```python
scale_params = learn_scaling_params(x_temp, SCALING_METHODS)
scale_params
```

```python
def scale_numerical(X, methods, params):
    X = X.copy()
    for var in methods:
        if methods[var] == 'normalize':
            X[var] = (X[var] - params[var]['min']) / (params[var]['max'] - params[var]['min'])
        elif methods[var] == 'standardize':
            X[var] = (X[var] - params[var]['mean']) / params[var]['stdev']
        else:
            raise ValueError("Methods can only be normalize and standardise.")
    return X
```

Now we perform one-hot encoding. Brand, model, and postcode have too many levels for one-hot encoding, and will create far too many binary features. So we are going to group all rare observations into a "rare" category. As before, we need to learn the top N categories in these three features from the training set.

```python
OH_FEATURES = [
    'vehicleType',
    'gearbox',
    'model',
    'monthOfRegistration',
    'fuelType',
    'brand',
    'postalCode'
]
OH_ENCODING_TOPS = dict(
    brand      = 10,
    model      = 20,
    postalCode = 30
)
```

```python
def learn_top_N(X, top_ns):
    top_n_cats = dict()
    for var, n in top_ns.items():
        top_n_cats[var] = X[var].value_counts()[:n].index.tolist()
    return top_n_cats
```

```python
learn_top_N(x_temp_scaled, OH_ENCODING_TOPS)
```

```python
from sklearn.preprocessing import OneHotEncoder

def onehot_encoder_fit(X, features, top_n = None):
    """
    Train the one-hot encoder, including finding the top N categories of high-cardinality
    categorical features.
    features: List of features to one-hot encode.
    top_n: dict, specifies the number of top categories to retain in the included categorical
    variables. The other categories will be replaced by "rare". If None, all categories of
    all categorical variables are one-hot encoded.
    :return: A trained one-hot encoder.
    """
    X = X.copy()
    # replace rare categories names with "rare"
    if top_n is not None:
        top_cats = learn_top_N(X, top_n)
        for var, top_values in top_cats.items():
            X[var] = np.where(X[var].isin(top_values), X[var], "rare")
            
    encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
    encoder.fit(X[features])
    return encoder

def onehot_encoder_transform(X, encoder, features, top_n = None):
    """
    Apply the learned `encoder` to X.
    features: list of categorical features to one-hot encode.
    :return: A dataframe where the specified features have been one-hot encoded.
    """
    X = X.copy()
    
    if top_n is not None:
        top_cats = learn_top_N(X, top_n)
        for var, top_values in top_cats.items():
            X[var] = np.where(X[var].isin(top_values), X[var], "rare")
    
    out = encoder.transform(X[features])

    try:
        feature_names_out = encoder.get_feature_names_out(features)
    except AttributeError:   # older versions of sklearn
        feature_names_out = encoder.get_feature_names(features)
    X_onehot = pd.DataFrame(
                out, 
                columns=feature_names_out,
                index=X[features].index
            )
    X = pd.concat([X, X_onehot], axis=1)
    X.drop(features, axis=1, inplace=True)
    return X
```

```python
# 1. Load and split the data and do some cleanup
cars_train, cars_val, cars_test = load_and_split('data/autos_eng.csv', 0.1, 0.1)
cars_train = remove_features(cars_train)
cars_train = cleanup(cars_train, 'train')
cars_val = remove_features(cars_val)
cars_val = cleanup(cars_val, 'valid')
cars_test = remove_features(cars_test)
cars_test = cleanup(cars_test, 'test')

# 2. Transform and clean labels, mark outliers as NAs
X_train, y_train = make_labels(cars_train, log_transform=True, label_bounds=(100, 100000))
X_train = outliers_to_nans(X_train, power_bounds=(25, 650), year_bounds=(1960, 2016), recent_km=10000)
X_val, y_val = make_labels(cars_val, log_transform=True, label_bounds=(100, 100000))
X_val = outliers_to_nans(X_val, power_bounds=(25, 650), year_bounds=(1960, 2016), recent_km=10000)
X_test, y_test = make_labels(cars_test, log_transform=True, label_bounds=(100, 100000))
X_test = outliers_to_nans(X_test, power_bounds=(25, 650), year_bounds=(1960, 2016), recent_km=10000)

# 3. Feature Engineering
X_train = add_vintage_interaction(X_train, 1996, True)
X_val = add_vintage_interaction(X_val, 1996, True)
X_test = add_vintage_interaction(X_test, 1996, True)

# 4. Postcode engineering for our baseline model
X_tr_base = transform_postcode(X_train)
X_val_base = transform_postcode(X_val)
X_tst_base = transform_postcode(X_test)

# 5. Imputation of missing values, scaling, encoding of categorical variables
impute_values = learn_impute_values(X_tr_base, NA_METHODS)
X_tr_base = impute_NA(X_tr_base, impute_values, na_flag=True)
X_val_base = impute_NA(X_val_base, impute_values, na_flag=True)
X_tst_base = impute_NA(X_tst_base, impute_values, na_flag=True)

scale_params = learn_scaling_params(X_tr_base, SCALING_METHODS)
X_tr_base = scale_numerical(X_tr_base, SCALING_METHODS, scale_params)
X_val_base = scale_numerical(X_val_base, SCALING_METHODS, scale_params)
X_tst_base = scale_numerical(X_tst_base, SCALING_METHODS, scale_params)

encoder = onehot_encoder_fit(X_tr_base, OH_FEATURES, OH_ENCODING_TOPS)
X_tr_base = onehot_encoder_transform(X_tr_base, encoder, OH_FEATURES, OH_ENCODING_TOPS)
X_val_base = onehot_encoder_transform(X_val_base, encoder, OH_FEATURES, OH_ENCODING_TOPS)
X_tst_base = onehot_encoder_transform(X_tst_base, encoder, OH_FEATURES, OH_ENCODING_TOPS)
```

We can now fit our model. Since we have a lot of features, we will select LASSO regression (linear regression with L1 regularisation). ### Geographical Features
We saw that postcode could help us import external geographical data. Let's do this now. Because the coordinate bins need to be learned from the training set and applied to the other two splits, we must remember their cut values. This is achieved by using a class. Since we would like to be able to include this preprocessing step into a sklearn pipeline, we will create the class as a subclass of sklearn's TransformerMixin.

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class CoordinatesFeature(BaseEstimator, TransformerMixin):
    def __init__(self, n_bins = 20, keep_postcode = False,  keep_coord = False):
        # bins: Number of bins to create in both longitude and latitude.
        # keep_orig_cood: Whether to keep the exact latitude and longitude values alongside
        # the binned values.
        super().__init__()
        self.postcodes = pd.read_csv(
            "data/plz_geocoord.csv",
            names=['postcode', 'lat', 'long'],
            header=0,
            dtype={'postcode': 'object'}
        )
        self.n_bins = n_bins
        self.bins = {'lat': 0, 'long': 0}
        self.keep_postcode = keep_postcode
        self.keep_coord = keep_coord
        self.label_avg = None
        self.feature_names_out = None

    def fit(self, X, y = None):
        X_temp = X.copy()
        X_temp = pd.merge(X_temp, self.postcodes, left_on='postalCode', right_on='postcode', how='left')
        lat_binned, self.bins['lat'] = pd.cut(X_temp['lat'], self.n_bins, retbins=True)
        long_binned, self.bins['long'] = pd.cut(X_temp['long'], self.n_bins, retbins=True)
        coord_codes = lat_binned.cat.codes.astype(str).str.cat(long_binned.cat.codes.astype(str), sep=',')
        coord_codes.name = 'coord_codes'
        
        return self
        
    def transform(self, X):
        X = X.copy()
        X = pd.merge(X, self.postcodes, left_on='postalCode', right_on='postcode', how='left')
        X.drop('postcode', axis=1, inplace=True)
        lat_binned = pd.cut(X['lat'], self.bins['lat'], labels=[str(n) for n in range(self.n_bins)])
        long_binned = pd.cut(X['long'], self.bins['long'], labels=[str(n) for n in range(self.n_bins)])   
        X['coord_codes'] = lat_binned.cat.codes.astype(str).str.cat(long_binned.cat.codes.astype(str), sep=',')

        if not self.keep_coord:
            X.drop(['lat', 'long'], axis=1, inplace=True)
        if not self.keep_postcode:
            X.drop('postalCode', axis=1, inplace=True)
        self.feature_names_out = X.columns.tolist()
        return X
    
    def get_feature_names_out(self):
        return self.feature_names_out
```

For missing value imputation, we will again subclass sklearn's TransformerMixin class to have the ability to learn the imputation values from the training set and apply them to all sets. To offer maximum flexibility, we will reuse the configuration dictionary created earlier, which will allow us to easily apply different NA imputation methods to each variable. We could also add KNN as an imputation method, but this needs to be done after categorical encoding because KNN can only learn from numerical values.

```python
NA_METHODS = dict(
    vehicleType         ='mode',
    yearOfRegistration  ='median',
    gearbox             ='mode',
    powerPS             ='median',
    model               ='mode',
    kilometer           ='median',
    monthOfRegistration ='mode',
    fuelType            ='mode',
    brand               ='mode',
    notRepairedDamage   ='mode',
    coord_codes         ='mode'   
)
```

```python
class NAHandler(BaseEstimator, TransformerMixin):
    def __init__(self, methods: dict, zero_string: str = '00', na_flag: bool = True):
        """
        methods: dict mapping each variable to the NA handling method. Each value must be one 
        of 'drop', 'mean', 'median', 'mode', 'zero', . Unspecified variables get 'mode' by default
        if they are categorical, 'median' if they are numerical, and 'zero' if they are strings. 
        zero_string: string to use when imputing a string variable with method 'zero'.
        na_flag: bool, set to True (default) if you want an extra binary column for each variable
        indicating rows where an NA imputation occurred.
        
        """
        super().__init__()
        # Check that all passed methods are valid
        accepted_methods = ['mean', 'median', 'mode', 'zero']
        assert all([v in accepted_methods for v in methods.values()])
        self.methods = methods
        self.zero_string = zero_string
        self.na_flag = na_flag
        self.values = dict()
        self.na_flags = list()
#         if 'knn' in self.methods.values():
#             from sklearn.impute import KNNImputer
#             self.knn_imputer = KNNImputer(missing_values='@#~|')  # use some unlikely string
        self.feature_names_out = None
            
    def fit(self, df, y = None):
        """
        Learns the imputation values for each column from the data in `df` and the options
        passed in the __init__ method.
        Warning: Do not split the features from the labels before using this function!
        """
        for var in df.columns:  
            # first ensure we capture all columns and that the methods make sense
            # and specify default methods for columns that are not passed in methods:
            if var not in self.methods:
                if df[var].dtypes == 'category': 
                    self.methods[var] = 'mode'
                elif df[var].dtypes == 'object':
                    self.methods[var] = 'zero'
                elif df[var].dtypes in ['float32', 'float64', 'int32', 'int64']: 
                    self.methods[var] = 'median'

        for var, method in self.methods.items():      
            if (df[var].isnull().sum() >= 1) and (self.na_flag):
                # if user asked for na flags, remember the list of added columns here
                self.na_flags.append(var)
        
            if method == 'mode':
                self.values[var] = df[var].mode()[0]
            elif method == 'median':
                assert df[var].dtypes in ['float32', 'float64', 'int32', 'int64'], "Wrong variable data type for `median`."
                self.values[var] = df[var].median()
            elif method == 'mean':
                assert df[var].dtypes in ['float32', 'float64', 'int32', 'int64'], "Wrong variable data type for `mean`."
                self.values[var] = df[var].mean()
            elif method == 'zero':
                assert df[var].dtypes != 'category', "Wrong variable data type for `zero`."
                if df[var].dtypes == 'object':
                    self.values[var] = self.zero_string
                else:
                    self.values[var] = 0
#             elif method == 'knn':
#                 self.values[var] = '@#~|'
#                 df_temp = df.copy()
#                 # set missing values to something that we can easily find and impute with KNN
#                 if df_temp[var].dtype == 'category':
#                     df_temp[var] = df_temp[var].cat.add_categories('@#~|')
#                 df_temp[var].fillna('@#~|', inplace=True)
            else:
                raise ValueError("Invalid method.")
#         if 'knn' in self.methods.values():
#             print(df_temp)
#             self.knn_imputer.fit(df_temp)
        return self
    
    def transform(self, df: pd.DataFrame):
        """
        Transforms `df` in line with the methods passed in the __init__ method and the
        imputation values learned in the fit method. Returns a transformed copy of `df`.
        Warning: Do not split the features from the labels before using this function!
        """
        df = df.copy()  # we don't want any side effects
        for var, value in self.values.items():
            if (self.na_flag) and (var in self.na_flags):
                # create the NA flag columns
                df[var + "_NA"] = np.where(df[var].isnull(), 1, 0)
            df[var].fillna(value, inplace=True)
#         if 'knn' in self.methods.values(): 
#             df[var] = self.knn_imputer.transform(df)
#             if df[var].dtype == 'category':
#                 df[var] = df[var].cat.remove_categories('@#~|')
        self.feature_names_out = df.columns.tolist()
        return df
    
    def get_feature_names_out(self):
        return self.feature_names_out
```

### Categorical Encoding
Now we are moving to encoding of variables. In EDA, we found that `brand` and `model` both have a similar issue: Because they so many different categories ("levels") each, they generate a huge number of binary columns when applying one-hot encoding to them. A possible solution was to group all rare categories into a "Rare" label, but it turned out that most of the observations would end up in this category. Let's try something else here: Target encoding. From the training, we will learn the mean price of each category (for each of these two variables), and replace the categorical variables by these mean values. This allows us to have only one column for each, while including important information. Of course, these mean prices are learned from the training set, and will be applied without modification on the validation and test sets.

This method can also be applied to the geographical bins created above, where each bin is replaced by its mean vehicle price.

Because we need to store these values, the easiest way is to subclass sklearn's TransformerMixin, which will also allow us to include the fitting and transformation into a sklearn pipeline.

```python
class TargetEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, features):
        super().__init__()
        # features: list of column names
        self.features = features
        self.averages = dict()
        self.feature_names_out = None
        
    def fit(self, X, y):
        X, y = X.copy(), y.copy()
        temp_df = pd.concat([X[self.features], y], axis=1)
        temp_df.columns = self.features + ['label']
        # Compute average of label for each level of the categorical variables in self.features:
        for var in self.features:
            self.averages[var] = temp_df.groupby(var).mean()['label']
            self.averages[var].name = var + '_mean'
        return self
    
    def transform(self, X):
        X = X.copy()
        for var in self.features:
            # Get the mean value of label for each value of the categorical variable
            X = pd.merge(X, self.averages[var], left_on=var, right_on=var, how='left')
            X.drop(var, axis=1, inplace=True)
        self.feature_names_out = X.columns.tolist()
        return X
        
    def get_feature_names_out(self):
        return self.feature_names_out
```

The other categorical variables will be simply encoded with one-hot encoding. We did it using functions when building our baseline model, but if we want to be able to enclose everything in a pipeline to fit and transform just once, we need to write another class to handle all encoding operations. To tell the pipeline which encoding to apply to which variables, we create a configuration dictionary.

```python
ENCODING_METHODS = dict(
    vehicleType         = 'onehot',
    gearbox             = 'onehot',
    fuelType            = 'onehot',
    monthOfRegistration = 'onehot',
    brand               = 'target',
    model               = 'target',
    coord_codes         = 'target',
)
```

```python
from sklearn.preprocessing import OneHotEncoder

class FeatureEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, methods):
        super().__init__()
        self.methods = methods
        self.onehot_vars = [var for var in methods if methods[var] == 'onehot']
        self.targetenc_vars = [var for var in methods if methods[var] == 'target']
        self.onehot_enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
        self.target_enc = TargetEncoder(self.targetenc_vars)
        self.feature_names_out = None
        
    def fit(self, X, y):
        X =  X.copy()
        if len(self.onehot_vars) > 0:
            X_onehot = X[self.onehot_vars].copy()
            self.onehot_enc.fit(X_onehot)
        if len(self.targetenc_vars) > 0:
            self.target_enc.fit(X, y)
        return self
    
    def transform(self, X):
        if len(self.onehot_vars) > 0:
            X_onehot_in = X[self.onehot_vars].copy()
            onehot_out = self.onehot_enc.transform(X_onehot_in)
            try:
                feature_names = self.onehot_enc.get_feature_names_out(
                    self.onehot_vars
                )
            except AttributeError:  # older version of sklearn
                feature_names = self.onehot_enc.get_feature_names(
                    self.onehot_vars
                )
            X_onehot_out = pd.DataFrame(
                onehot_out, 
                columns=feature_names,
                index=X_onehot_in.index
            )
        if len(self.targetenc_vars) > 0:
            # Don't target-encode variables that have been one-hot-encoded
            X_target_in = X.copy().drop(self.onehot_vars, axis=1)
            X_target_out = self.target_enc.transform(X_target_in)
        
        if (len(self.onehot_vars) > 0) & (len(self.targetenc_vars) > 0):
            X = pd.concat([X_target_out, X_onehot_out], axis=1)
            self.feature_names_out = X.columns.tolist()
            return X
        elif len(self.onehot_vars) > 0:
            X = pd.concat([X, X_onehot_out], axis=1)
            self.feature_names_out = X.columns.tolist()
            
            return X 
        else:
            self.feature_names_out = X_target_out.columns.tolist()
            return X_target_out
        
    def get_feature_names_out(self):
        return self.feature_names_out
```

### Feature Scaling

All that remains to do now is to scale numerical values.

```python
SCALING_METHODS = dict(
    yearOfRegistration   = 'normalize',
    powerPS              = 'standardize',
    kilometer            = 'normalize',
#     monthOfRegistration  = 'normalize',
)
```

```python
class FeatureScaler(BaseEstimator, TransformerMixin):
    def __init__(self, methods):
        super().__init__()
        self.params = dict()
        self.methods = methods
        self.feature_names_out = None
        
    def fit(self, X, y=None):
        for var, method in self.methods.items():
            if method == 'normalize':
                self.params[var] = {'min': X[var].min(), 'max': X[var].max()}
            elif method == 'standardize':
                self.params[var] = {'mean': X[var].mean(), 'stdev': X[var].std()}
            else:
                raise ValueError("Methods can only be normalize and standardise.")
        return self
    
    def transform(self, X):
        X = X.copy()
        for var in self.methods:
            if self.methods[var] == 'normalize':
                X[var] = (X[var] - self.params[var]['min']) / (
                    self.params[var]['max'] - self.params[var]['min'])
            elif self.methods[var] == 'standardize':
                X[var] = (X[var] - self.params[var]['mean']) / self.params[var]['stdev']
            else:
                raise ValueError("Methods can only be normalize and standardise.")
        self.feature_names_out = X.columns.tolist()
        return X    
    
    def get_feature_names_out(self):
        return self.feature_names_out
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor

preprocess = make_pipeline(
        CoordinatesFeature(n_bins=20, keep_postcode=False, keep_coord=False),
        NAHandler(NA_METHODS, na_flag=True),
        FeatureEncoder(ENCODING_METHODS),
        FeatureScaler(SCALING_METHODS)
)

rf_reg = RandomForestRegressor(n_estimators=300, criterion='mse', max_features=0.2, n_jobs=-1)
```
