## Practical Example
### Non-Modular Code

```python
from sklearn.model_selection import train_test_split
pd.set_option('mode.chained_assignment', None)

cars = pd.read_csv(
    "data/autos_clean.csv",
    parse_dates=[0, 12, 14],
    dtype={
        'vehicleType': 'category',
        'monthOfRegistration': 'category',
        'gearbox': 'category',
        'model': 'category',
        'fuelType': 'category',
        'brand': 'category',
        'notRepairedDamage': 'category',
        'postalCode': 'str'
    }
)
# Drop datetime variables because they won't be available for the final application
cars.drop(['dateCrawled', 'dateCreated', 'lastSeen'], axis='columns', inplace=True)

# Assign order to monthOfRegistration
cars.monthOfRegistration = cars.monthOfRegistration.cat.reorder_categories(
    [str(s) for s in range(1, 13)],
    ordered=True
)

# Select variables by type
target = cars.price.name
cat_vars = cars.select_dtypes('category').columns
num_vars = cars.select_dtypes(['float', 'int']).columns.drop('price')
str_vars = cars.select_dtypes('object').columns
# Note: We ignore datetime variables because we are going to drop them later

# Split training and test sets
cars_tr, cars_tst = train_test_split(cars, test_size=0.1, random_state=0)

# Duplicate removal
cars_tr.drop_duplicates(inplace=True)

# NA imputation
for var in cat_vars:  # categorical variables: use mode
    cars_tr[var + '_NA'] = np.where(cars_tr[var].isnull(), 1, 0)
    cars_tr[var].fillna(cars_tr[var].mode()[0], inplace=True)
    # Apply same transformations to test set, using value from train set
    cars_tst[var + '_NA'] = np.where(cars_tst[var].isnull(), 1, 0)
    cars_tst[var].fillna(cars_tr[var].mode()[0], inplace=True)

for var in num_vars:  # numeric variables: use median
    cars_tr[var + '_NA'] = np.where(cars_tr[var].isnull(), 1, 0)
    cars_tr[var].fillna(cars_tr[var].median(), inplace=True)
    # Apply same transformations to test set, using value from train set
    cars_tst[var + '_NA'] = np.where(cars_tst[var].isnull(), 1, 0)
    cars_tst[var].fillna(cars_tr[var].median(), inplace=True)

for var in str_vars:  # string variables: create label for unknown
    cars_tr[var].fillna('00000', inplace=True)
    cars_tst[var].fillna('00000', inplace=True)
    
    
# Outlier clipping
for var in num_vars:
    lower, upper = (cars_tr[var].mean() - 3 * cars_tr[var].std(), 
                    cars_tr[var].mean() + 3 * cars_tr[var].std())
    cars_tr[var + 'OUT'] = np.where(
        cars_tr[var] < lower, -1,
        np.where(
            cars_tr[var] > upper, 1,
            0
        )
    )
    cars_tr[var].clip(lower=lower, upper=upper, inplace=True)
    
    # Fix test set too using values learned from train set
    cars_tst[var + 'OUT'] = np.where(
        cars_tst[var] < lower, -1,
        np.where(
            cars_tst[var] > upper, 1,
            0
        )
    )
    cars_tst[var].clip(lower=lower, upper=upper, inplace=True)

# Rare categorical labels
for var in cat_vars:
    labels_by_freq = cars_tr.groupby(var).size() / len(cars_tr)
    # Labels are considered as rare when they appear in less than 0.5% of the observations
    rare_labels = labels_by_freq[labels_by_freq < 0.005].index.tolist()
    print(f"Variable {var}: Found rare labels {rare_labels}")
    # We also need to remember all labels seen in the train set, in case the test sest
    # has unseen labels (which we will then also replace by "Rare")
    all_train_labels = cars_tr[var].unique().categories.tolist()
    # Learn from the training set, apply to both data sets:
    cars_tr[var] = np.where(cars_tr[var].isin(rare_labels), "Rare", cars_tr[var])
    cars_tst[var] = np.where(cars_tst[var].isin(rare_labels), "Rare", cars_tst[var])
    # Unseen labels in the test set:
    cars_tst[var] = np.where(cars_tst[var].isin(all_train_labels), cars_tst[var], "Rare")    
```

### Modular Code

```python
from typing import Union
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
na_methods = dict(
    price              ='drop',  # no point imputing missing labels
    vehicleType        ='mode',
    yearOfRegistration ='median',
    gearbox            ='mode',
    powerPS            ='median',
    model              ='mode',
    kilometer          ='median',
    monthOfRegistration='mode',
    fuelType           ='mode',
    brand              ='mode',
    notRepairedDamage  ='mode',
    postalCode         ='zero'   
)
outlier_methods = dict(
    price              ='drop',  # no point fixing outlier labels
    powerPS            ='median',
    yearOfRegistration ='median',
    model              ='group',
    brand              ='group',
)
outlier_thresholds = dict(
    price              =(100, 100000),
    powerPS            =(25, 650),
    yearOfRegistration =(1960, 2016),
    model              =0.003,  # minimum frequency to have its own label
    brand              =0.01,  # minimum frequency to have its own label
)
```

```python
def create_datasets(
    pathname: str,
    date_col_pos: list[int] = [0, 12, 14], 
    cat_vars: list[str] = [
        'vehicleType', 'monthOfRegistration', 'gearbox', 'model',
         'fuelType', 'brand', 'notRepairedDamage'
    ],
    postcode_as_str: bool = True,
    test_size: float = 0.10,
    seed: int = 0
) -> tuple[pd.DataFrame]:
    """
    Returns a training set and a test set by loading the file located in `pathname`, using 
    test set size `test_size` and the loading options passed in `date_col_pos`, `cat_vars`,
    and `postcode_as_str`. Set the `seed` manually for reproducibility of the train/test 
    splitting operation.
    """
    assert not (('postalCode' in cat_vars) and (postcode_as_str)), \
        "Passed `postalCode` as categorical and `postcode_as_str` as True, which is inconsistent."
    assert 0 < test_size < 1, "Test set size cannot be (<=0 or >= 1) * total data length"
    
    dtypes = {var: 'category' for var in cat_vars}
    if postcode_as_str:
        dtypes['postalCode'] = 'str'
    df = pd.read_csv(pathname, parse_dates=date_col_pos, dtype=dtypes)
    df.drop(['dateCrawled', 'dateCreated', 'lastSeen'], axis='columns', inplace=True)
    # Assign order to `monthOfRegistration`
    if 'monthOfRegistration' in cat_vars:
        df['monthOfRegistration'] = df['monthOfRegistration'].cat.reorder_categories(
            [str(s) for s in range(1, 13)],
            ordered=True
    )
    return train_test_split(df, test_size=test_size, random_state=seed)
```

```python
def update_variable_types(df: pd.DataFrame):
    """
    Returns a dict providing the column indexes for each of categorical, numerical, and 
    string variables in `df`. This is useful for updating these lists after data 
    transformations that may alter the type of some variables.
    """
    return dict(
        cat_vars = df.select_dtypes('category').columns,
        num_vars = df.select_dtypes(['float', 'int']).columns,
        str_vars = df.select_dtypes('object').columns
    )
```

```python
def remove_duplicates(df: pd.DataFrame) -> pd.DataFrame:
    return df.drop_duplicates()
```

```python
class NAHandler:
    def __init__(self, methods: dict, zero_string: str = '00000', na_flag: bool = True):
        """
        methods: dict mapping each variable to the NA handling method. Each value must be one 
        of 'drop', 'mean', 'median', 'mode', or 'zero'. Unspecified variables get 'mode' by default
        if they are categorical, 'median' if they are numerical, and 'zero' if they are strings. 
        zero_string: string to use when imputing a string variable with method 'zero'.
        na_flag: bool, set to True (default) if you want an extra binary column for each variable
        indicating rows where an NA imputation occurred.
        
        """
        # Check that all passed methods are valid
        accepted_methods = ['drop', 'mean', 'median', 'mode', 'zero']
        assert all([v in accepted_methods for v in methods.values()])
        self.methods = methods
        self.zero_string = zero_string
        self.values = dict()
        self.na_flag = na_flag
        self.na_flags = list()
        
    def fit(self, df: pd.DataFrame):       
        """
        Learns the imputation values for each column from the data in `df` and the options
        passed in the __init__ method.
        Warning: Do not split the features from the labels before using this function!
        """
        for var in df.columns:  
            # first ensure we capture all columns and that the methods make sense
            # and specify default methods for columns that are not passed in methods:
            if var not in self.methods:
                if df[var].dtypes == 'category': 
                    self.methods[var] = 'mode'
                elif df[var].dtypes in ['float64', 'int64']: 
                    self.methods[var] = 'median'
                else: 
                    self.methods[var] = 'zero'
        
        for var, method in self.methods.items():
            if method == 'drop':
                self.values[var] = None
                continue
        
            if (df[var].isnull().sum() >= 1) and (self.na_flag):
                # if user asked for na flags, remember the list of added columns here
                self.na_flags.append(var)
        
            if method == 'mode':
                self.values[var] = df[var].mode()[0]
            elif method == 'median':
                assert df[var].dtypes in ['float64', 'int64'], "Wrong variable data type for `median`."
                self.values[var] = df[var].median()
            elif method == 'mean':
                assert df[var].dtypes in ['float64', 'int64'], "Wrong variable data type for `mean`."
                self.values[var] = df[var].mean()
            elif method == 'zero':
                assert df[var].dtypes != 'category', "Wrong variable data type for `zero`."
                if df[var].dtypes == 'object':
                    self.values[var] = self.zero_string
                else:
                    self.values[var] = 0
            else:
                raise ValueError("Invalid method.")
    
    def transform(self, df: pd.DataFrame):
        """
        Transforms `df` in line with the methods passed in the __init__ method and the
        imputation values learned in the fit method. Returns a transformed copy of `df`.
        Warning: Do not split the features from the labels before using this function!
        """
        df = df.copy()  # we don't want any side effects
        for var, value in self.values.items():
            if value is None:
                df.dropna(axis=0, subset=[var], inplace=True)
            else:
                if (self.na_flag) and (var in self.na_flags):
                    # create the NA flag columns
                    df[var + "_NA"] = np.where(df[var].isnull(), 1, 0)
                df[var].fillna(value, inplace=True)
        return df
```

```python
class GeneralOutlierHandler:
    def __init__(
        self, 
        methods: dict,
        thresholds: dict[Union[tuple, float]], 
        outlier_flag: bool = True
    ):
        """
        methods: dict mapping columns with a treatment method for outliers. The dict values 
        can be 'drop', 'median', 'mean', 'clip' for numerical variables, 'drop', mode', 'group'
        for categorical variables.
        thresholds: dict mapping columns with value thresholds for what is considered an 
        outlier. The values of the dict are (lower threshold, upper threshold) tuples for 
        numerical variables, and a lower threshold float for categorical variables if they
        are treated with the `group` method.
        
        methods and threshold must have the same length. 
        The `group` method groups all categorical labels with a frequency less than the
        specified threshold into a "Rare" class.
        Variables other than numerical and categorical variables in the methods and thresholds dicts
        (e.g. string or datetime variables) will be ignored.
        """
        assert [(m in ['drop', 'median', 'mean', 'clip'] and isinstance(t, tuple)) or
               (m in ['drop', 'mode', 'group'] and isinstance(t, float))
                for m, t in zip(methods, thresholds)], \
                "Inconsistent or erroneous arguments. Please refer to the docstring."
                
        self.methods = methods
        self.thresholds = thresholds
        self.outlier_flag = outlier_flag
        self.flags = list()
        self.replace_values = dict()  # for vars using median, mean, or mode
        self.rare_labels = dict()  # for categorical vars using group
        self.train_labels = dict()  # remember all labels seen during training
        
    def fit(self, df: pd.DataFrame):
        """
        Learns from `df` the columns that contain outliers, and the values to replace
        them with if required, according to the criteria passed to the __init__ method.
        """
        for var, method in self.methods.items():
            if df[var].dtypes in ['float64', 'int64']:
                # ensure consistency of methods and variable types
                assert method in ['drop', 'median', 'mean', 'clip'], \
                    f"Inconsistent method {method} passed against variable {var}."
                
                lower, upper = self.thresholds[var][0], self.thresholds[var][1]
                if (len(df.loc[(df[var] < lower) | (df[var] > upper)]) > 0):
                    if self.outlier_flag:
                        self.flags.append(var)              
                    if method == 'median':
                        self.replace_values[var] = df[var].median()
                    elif method == 'mean':
                        self.replace_values[var] = df[var].mean()
                
            elif df[var].dtypes == 'category':
                assert method in ['drop', 'mode', 'group'], \
                    f"Inconsistent method {method} passed against variable {var}."
                
                self.train_labels[var] = df[var].unique().tolist()
                labels_by_freq = df.groupby(var).size() / len(cars_tr)
                rare = labels_by_freq[
                    labels_by_freq < self.thresholds[var]
                ].index.tolist()  # collect rare labels
                if len(rare) > 0:
                    self.rare_labels[var] = rare
                    if self.outlier_flag:
                        self.flags.append(var)
                    if method == 'mode':
                        self.replace_values[var] = df[var].mode()[0]
    
    def transform(self, df: pd.DataFrame):
        df = df.copy()
        for var, method in self.methods.items():
            if df[var].dtypes == 'category':
                if method == 'drop':
                    df = df[~df[var].isin(self.rare_labels[var])]
                elif method == 'mode':
                    if (self.outlier_flag) and (var in self.flags):
                        # create flag column
                        df[var + "_OUT"] = np.where(
                            df[var].isin(self.rare_labels[var]), 
                            1, 0
                        )
                    df[var] = np.where(
                        df[var].isin(self.rare_labels[var]), 
                        self.replace_values[var], 
                        df[var]
                    )
                elif method == 'group':
                    if (self.outlier_flag) and (var in self.flags):
                        # create flag column
                        df[var + "_OUT"] = np.where(
                            df[var].isin(self.rare_labels[var]), 
                            1, 0
                        )
                    df[var] = np.where(
                        df[var].isin(self.rare_labels[var]), 
                        "Rare", 
                        df[var]
                    )
                    # Also replace any labels not seen in the training set by "Rare"
                    df[var] = np.where(
                        ~df[var].isin(self.train_labels[var]),
                        "Rare",
                        df[var]
                    )
                    
            elif df[var].dtypes in ['float64', 'int64']:
                lower, upper = self.thresholds[var][0], self.thresholds[var][1]
                if method == 'drop':
                    df = df[(df[var] >= lower) & (df[var] <= upper)]
                elif method == 'clip':
                    if (self.outlier_flag) and (var in self.flags):
                        # create flag column
                        df[var + "_OUT"] = np.where(
                             (df[var] < lower) | (df[var] > upper), 
                            1, 0
                        )
                    df[var].clip(lower=lower, upper=upper, inplace=True)
                else:  # method == 'median' or 'mean'
                    if (self.outlier_flag) and (var in self.flags):
                        # create flag column
                        df[var + "_OUT"] = np.where(
                             (df[var] < lower) | (df[var] > upper), 
                            1, 0
                        )
                    df[var] = np.where(
                        (df[var] < lower) | (df[var] > upper),
                        self.replace_values[var],
                        df[var]
                    )
                    
        return df
```

```python
def sanitise(
    input_filepath: str, 
    na_methods: dict,
    outlier_methods: dict,
    outlier_thresholds: dict,
    save: bool = True):
    
    # split dataset and remove duplicates
    df_tr, df_tst = create_datasets(input_filepath)
    df_tr, df_tst = remove_duplicates(df_tr), remove_duplicates(df_tst)
    
    # Handle NAs
    na_handler = NAHandler(methods=na_methods, zero_string='00000', na_flag=True)
    # learn on training set, transform both
    na_handler.fit(df_tr)
    df_tr, df_tst = na_handler.transform(df_tr), na_handler.transform(df_tst)
    
    # Handle outliers
    gen_outlier_handler = GeneralOutlierHandler(
        methods=outlier_methods, 
        thresholds=outlier_thresholds,
        outlier_flag=True
    )
    # learn on training set, transform both
    gen_outlier_handler.fit(df_tr)
    df_tr, df_tst = gen_outlier_handler.transform(df_tr), gen_outlier_handler.transform(df_tst)
    
    if save:
        # save both files
        base_filepath = '.'.join(input_filepath.split('.')[:-1])  # get everything before final "."
        tr_filepath = base_filepath + "_train.csv"
        tst_filepath = base_filepath + "_test.csv"
        df_tr.to_csv(tr_filepath, index=False)
        df_tst.to_csv(tst_filepath, index=False)
    
    return df_tr, df_tst
```

```python
df_tr, df_tst = sanitise("data/autos_clean.csv", na_methods, outlier_methods, outlier_thresholds)
```
